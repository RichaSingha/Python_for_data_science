```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
mov = pd.read_csv("C:\\Users\\Richa 1\\Desktop\\Deskop\\python programming\\movies-Data.csv", encoding='latin1')
```

```python
mov.head()
```

```python
filter_genre = ['action','adventure','animation','comedy','drama']
mov = mov[mov.Genre.isin(filter_genre)]
```

```python
filter_studio = ['Buena Vista Studios','Fox','Paramount Pictures','Sony','Universal','WB']
mov = mov[mov.Studio.isin(filter_studio)]
```

```python
mov.Studio.unique()
```

```python
mov.Genre.unique()
```

```python
sns.set(style='darkgrid', palette='muted', color_codes=True)
```

```python
vis1 = sns.boxplot(data=mov, x='Genre', y='Gross % US', orient='v', color='lightgray', showfliers=False)
plt.setp(vis1.artists, alpha=0.5)
sns.stripplot(data=mov, x='Genre', y='Gross % US', jitter=True, size=6, linewidth=0, hue='Studio', alpha=0.7)

vis1.axes.set_title('Domestic Gross % by Genre', fontsize=30)
vis1.set_xlabel('Genre', fontsize=20)
vis1.set_ylabel('Gross % US', fontsize=20)
vis1.legend(loc=2, bbox_to_anchor=(1,1))

plt.show()
```
